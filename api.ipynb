```python
import pandas as pd
import requests
```

```python
url = "https://api.coingecko.com/api/v3/coins/list"
response = requests.get(url)
data = response.json()
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv("coingecko_coin_list.csv", index=False)
    
```
